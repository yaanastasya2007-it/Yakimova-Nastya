```python
import pygame
import random
import math

pygame.init()
screen = pygame.display.set_mode((1300, 900), pygame.RESIZABLE)
pygame.display.set_caption("Солнечная система")
pygame.mixer.init()
pygame.mixer.music.load("neon.wav")
pygame.mixer.music.play(-1) 

WIDTH, HEIGHT = 1300, 900
cx = WIDTH // 2
cy = HEIGHT // 2   
FPS = 60
clock = pygame.time.Clock()
a,b=1300,0

comete_image = pygame.image.load("comete.jpg").convert()
comete_image = pygame.transform.scale(comete_image, (40,40))
comete_image.set_colorkey((255, 255, 255))

comete_x = random.randint(40, 1300 - 40)
comete_y = random.randint(40, 900 - 40)

class Planet:
    def __init__(self, screen, radius, orbit_radius, color=None, speed=0, angle=0, image_path=None):
        self.screen = screen
        self.radius = radius
        self.orbit_radius = orbit_radius
        self.color = color
        self.speed = speed
        self.angle = angle
        self.x = 0
        self.y = 0
        self.image = None
        if image_path:
        
            self.image = pygame.image.load(image_path).convert_alpha()
            self.image = pygame.transform.scale(self.image, (radius*2, radius*2))
            self.image.set_colorkey((255, 255, 255))


    def update(self, dt):
        global cx, cy
        self.angle += self.speed * dt
        self.x = cx + self.orbit_radius * math.cos(self.angle)
        self.y = cy + self.orbit_radius * math.sin(self.angle)

    def draw(self):
        if self.image:
            self.screen.blit(self.image, (int(self.x - self.radius), int(self.y - self.radius)))
        else:
            pygame.draw.circle(self.screen, self.color, (int(self.x), int(self.y)), self.radius)
        
class Moon:
    def __init__(self, screen, radius_1, orbit_radius_1, color_1=None, speed_1=0, angle_1=0, image_1=None):
        self.screen = screen
        self.radius_1 = radius_1
        self.orbit_radius_1 = orbit_radius_1
        self.color_1 = color_1
        self.speed_1 = speed_1
        self.angle_1 = angle_1
        self.x = 0
        self.y = 0
        self.image_1 = None
    def update(self, dt):
        global cx1, cy1
        self.angle_1 += self.speed_1 * dt
        self.x = cx1 + self.orbit_radius_1 * math.cos(self.angle_1)
        self.y = cy1 + self.orbit_radius_1 * math.sin(self.angle_1)

    def draw(self):
       pygame.draw.circle(self.screen, self.color_1, (int(self.x), int(self.y)), self.radius_1)
       
sun_color = (255, 200, 0)
sun_radius = 20

mercury = Planet(screen, radius=20, orbit_radius=50, speed=0.7, image_path="mercury.jpg")
venus = Planet(screen, radius=30, orbit_radius=100, speed=0.5, image_path="venus.jpg")
earth = Planet(screen, radius=40, orbit_radius=150, speed=0.2, image_path="earth1.jpg")
mars  = Planet(screen, radius=50, orbit_radius=200, speed=0.15, image_path="mars.jpg")
jupyter = Planet(screen, radius=80, orbit_radius=250, speed=0.1, image_path="jupyter.png")
saturn = Planet(screen, radius=60, orbit_radius=300, speed=0.09, image_path="saturn.jpg")
uran = Planet(screen, radius=75, orbit_radius=480, speed=0.05, image_path="uran.jpg")
neptun = Planet(screen, radius=75, orbit_radius=540, speed=0.01, image_path="neptun.jpg")
plyton = Planet(screen, radius=60, orbit_radius=600, speed=0.001, image_path="plyton.jpg")
#moon  = Moon(screen, radius_1=6, orbit_radius_1=180, color_1=(255, 100, 80), speed_1=0.6)

running = True
flag = True
while running:
    dt = clock.tick(FPS) / 1000.0
    #dt1 = clock.tick(FPS) / 1000.0
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.VIDEORESIZE:
            WIDTH, HEIGHT = event.w, event.h
            screen = pygame.display.set_mode((WIDTH, HEIGHT), pygame.RESIZABLE)
            cx = WIDTH // 2
            cy = HEIGHT // 2  
    mercury.update(dt)
    venus.update(dt)
    earth.update(dt)
    mars.update(dt)
    jupyter.update(dt)
    saturn.update(dt)
    uran.update(dt)
    neptun.update(dt)
    plyton.update(dt)
    #Moon.update(dt)
    screen.fill((0, 0, 20))
    pygame.draw.circle(screen, sun_color, (cx, cy), sun_radius)
    mercury.draw()
    venus.draw()
    earth.draw()
    #Moon.draw()
    mars.draw()
    jupyter.draw()
    saturn.draw()
    uran.draw()
    neptun.draw()
    plyton.draw()
    screen.blit(comete_image, (40, 40))
    if flag == True:
        comete_x = random.randint(40, 1300 - 40)
        comete_y = random.randint(40, 900 - 40)
        flag = False
    else:
        flag = True
    pygame.display.flip()

pygame.quit()
```
